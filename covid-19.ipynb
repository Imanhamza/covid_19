The total Recovered cases for each WHO Region, and which WHO Region have the most number?

```python
total_Recovered_df = df.groupby('WHO Region')['Recovered'].sum()
total_Recovered_df
```

The average death percentage for each WHO Region?

```python
df['deaths_precentage'] = df.Deaths / (df.Deaths + df.Active + df.Recovered)
df.columns
```

```python
a = df[df['WHO Region'] == 'Americas']['deaths_precentage'].mean()
print(f'Deaths percentage in Americas is {a:.2f}')
em =df[df['WHO Region'] == 'Eastern Mediterranean']['deaths_precentage'].mean()
print(f'Deaths percentage in Eastern Mediterranean is {em:.2f}')
eu = df[df['WHO Region'] == 'Europe']['deaths_precentage'].mean()
print(f'Deaths percentage in Europe is {eu:.2f}')
af = df[df['WHO Region'] == 'Africa']['deaths_precentage'].mean()
print(f'Deaths percentage in Africa is {af:.2f}')
sea = df[df['WHO Region'] == 'South-East Asia']['deaths_precentage'].mean()
print(f'Deaths percentage in South-East Asia is {sea:.2f}')
wp = df[df['WHO Region'] == 'Western Pacific']['deaths_precentage'].mean()
print(f'Deaths percentage in Western Pacific is {wp:.2f}')
```

```python
death_group = df.groupby('WHO Region')['deaths_precentage'].mean()
death_group
```

```python
plt.plot(death_group);
```
